## Papel Moeda em Circulação

```python
moeda_em_circulacao = sgs.get({'Moeda': 1780}, start='2002-01-01', end=date.today())
```

```python
moeda_em_circulacao.head()
```

```python
sns.set_theme(style="white")
fig, ax = plt.subplots(figsize=(18, 6))

ax = sns.lineplot(data=moeda_em_circulacao, x=moeda_em_circulacao.index , y='Moeda' ,ax=ax)
sns.despine(bottom=True, left=True)
plt.title("Moeda em Circulação", fontsize=18, pad=20)

ax.axvspan('2003-01-01', '2010-12-31', color='lightyellow', alpha=0.3, label='Governo Lula')
ax.axvspan('2011-01-01', '2016-12-31', color='lightcoral', alpha=0.3, label='Governo Dilma')
ax.axvspan('2017-01-01', '2018-12-31', color='lightgreen', alpha=0.3, label='Governo Temer')
ax.axvspan('2019-01-01', '2022-12-31', color='lightgrey', alpha=0.3, label='Governo Bolsonaro')
ax.axvspan('2020-02-28', '2022-04-22', color='red', alpha=0.3, label='Pandemia Covid-19')

plt.legend(loc='best')
plt.show()
```
