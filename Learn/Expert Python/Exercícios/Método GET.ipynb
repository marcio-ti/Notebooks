### Exercícios sobre o método `get()` em dicionários no Python

#### **Exercícios básicos**

1. **Acessando valores em um dicionário com `get()`**
   Crie um dicionário chamado `frutas` com as seguintes chaves e valores:
   ```python
   {"maçã": 3, "banana": 5, "laranja": 2}
   ```
   Use o método `get()` para acessar o número de bananas no dicionário. Em seguida, tente acessar uma chave chamada `"uva"` e defina um valor padrão como `0` para o caso de a chave não existir.

---

2. **Usando `get()` para evitar erros**
   Crie um dicionário chamado `estoque` com as seguintes informações:
   ```python
   {"caneta": 10, "caderno": 5, "borracha": 7}
   ```
   Escreva um código que verifique o estoque de um item chamado `"régua"`. Se o item não existir, retorne a mensagem: `"Item não encontrado"`.

---

3. **Valor padrão com `get()`**
   Crie um dicionário chamado `notas_alunos` que associa nomes de alunos às suas notas:
   ```python
   {"João": 8.5, "Maria": 9.0, "Pedro": 7.0}
   ```
   Use `get()` para buscar a nota de "Ana", retornando `0` caso ela não esteja no dicionário.

---

4. **Substituindo valores padrão**
   Crie um dicionário chamado `precos_produtos` com os preços de 3 produtos:
   ```python
   {"arroz": 20, "feijão": 15, "açúcar": 10}
   ```
   Escreva uma função chamada `obter_preco(produto)` que recebe o nome de um produto e retorna o preço dele usando `get()`. Se o produto não existir, retorne `-1`.

---

5. **Construindo dicionários com valores padrão**
   Escreva um código que verifica se uma chave está em um dicionário chamado `usuarios_online` e retorna um valor padrão `"Offline"` caso o usuário não esteja presente no dicionário. Use o método `get()`.

---

#### **Exercícios intermediários**

6. **Iteração e uso de `get()`**
   Dado um dicionário chamado `vendas_mensais` com as vendas de cada mês:
   ```python
   {"janeiro": 1000, "fevereiro": 1500, "março": 1200}
   ```
   Escreva um código que tente acessar as vendas de "abril" e retorne `0` caso o mês não esteja no dicionário. Adicione uma verificação que informe ao usuário se o valor retornado é o padrão ou não.

---

7. **Atualizando valores com `get()`**
   Crie um dicionário chamado `pontuacao` que associa jogadores aos seus pontos:
   ```python
   {"Alice": 50, "Bob": 80}
   ```
   Escreva uma função que recebe o nome de um jogador e a quantidade de pontos a ser adicionada. Use `get()` para obter a pontuação atual e atualize o valor no dicionário.

---

8. **Dicionário com listas e `get()`**
   Dado um dicionário chamado `carrinho_compras`, onde as chaves são categorias e os valores são listas de produtos:
   ```python
   {
       "frutas": ["maçã", "banana"],
       "verduras": ["alface", "rúcula"]
   }
   ```
   Escreva uma função chamada `adicionar_produto(categoria, produto)` que adiciona um novo produto à categoria correspondente no dicionário. Use `get()` para verificar se a categoria existe; caso contrário, crie uma nova categoria com o produto.

---

#### **Exercícios avançados**

9. **Substituindo chaves ausentes por valores padrão**
   Dado um dicionário chamado `desempenho` com os seguintes dados:
   ```python
   {"Ana": 8.5, "Carlos": 7.0, "Fernanda": 9.5}
   ```
   Escreva um código que recebe uma lista de alunos e, para cada aluno, verifica se ele está no dicionário. Se o aluno não estiver presente, atribua a nota padrão `6.0` e atualize o dicionário.

---

10. **Analisando dados com múltiplas verificações usando `get()`**
    Considere um dicionário chamado `resultados_prova` que contém os resultados de uma prova de alunos:
    ```python
    {
        "Matemática": {"João": 7.5, "Maria": 8.0},
        "Português": {"João": 6.0, "Pedro": 9.0}
    }
    ```
    Escreva uma função que recebe o nome de um aluno e a matéria, e retorna a nota correspondente. Se a matéria ou o aluno não existir no dicionário, retorne `"Nota não encontrada"`. Use `get()` para lidar com as verificações.

---

Esses exercícios cobrem desde o uso básico do método `get()` até aplicações mais complexas com dicionários aninhados e listas.
