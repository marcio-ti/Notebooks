**Trabalhando com Arquivos Excel usando openpyxl: Exercícios do Básico ao Avançado**

### **Nível Básico**

1. Crie uma nova planilha Excel chamada "dados.xlsx" e adicione o texto "Hello, World!" na célula A1.
2. Leia o valor da célula B2 de um arquivo Excel existente.
3. Adicione uma nova aba chamada "Planilha2" em um arquivo Excel.
4. Altere o nome da aba padrão de um arquivo Excel para "Relatório".
5. Salve um arquivo Excel após modificar o valor da célula C3 para "Python".
6. Escreva os números de 1 a 10 na coluna A de uma planilha.
7. Leia todos os valores da primeira linha de uma planilha e imprima-os.
8. Verifique se uma aba chamada "Resumo" existe em um arquivo Excel.
9. Formate a célula D4 para exibir os valores como moeda.
10. Altere a cor de fundo da célula E5 para amarelo.

### **Nível Intermediário**

11. Leia os valores de uma coluna inteira e armazene-os em uma lista.
12. Crie um arquivo Excel com três abas: "Vendas", "Estoque" e "Clientes".
13. Escreva uma função que receba um dicionário e escreva seus pares chave-valor em uma planilha.
14. Calcule a soma de todos os valores em uma coluna especificada por um usuário.
15. Leia uma tabela de dados de uma planilha e armazene-a em um DataFrame do pandas.
16. Apague uma aba chamada "Teste" de um arquivo Excel, caso ela exista.
17. Insira uma borda ao redor do intervalo A1:C3 em uma planilha.
18. Mescle as células de A1 até C1 e adicione o texto "Relatório Mensal".
19. Ajuste automaticamente a largura de todas as colunas de uma planilha baseada nos dados.
20. Salve cópias de um arquivo Excel em diferentes formatos, como .xlsx e .csv.

### **Nível Avançado**

21. Leia e imprima apenas os valores numéricos de uma planilha.
22. Crie uma função que valide se os dados em uma coluna estão ordenados em ordem crescente.
23. Aplique validação de dados para restringir uma coluna a aceitar apenas números entre 1 e 100.
24. Adicione um gráfico de linhas em uma planilha baseado nos valores de uma tabela.
25. Construa uma função que duplique o conteúdo de uma aba, criando uma cópia com um novo nome.
26. Extraia e imprima os nomes de todas as abas em um arquivo Excel.
27. Escreva uma função que destaque todas as células com valores negativos em vermelho.
28. Leia os dados de uma tabela em uma planilha e calcule a média dos valores em cada linha.
29. Crie uma macro simples que automatize a limpeza de dados em uma planilha.
30. Combine os dados de duas abas em uma terceira aba no mesmo arquivo Excel.

### **Nível Avançado com Outros Elementos**

#### **Listas e Dicionários**

31. Dada uma lista de listas, escreva cada sublista como uma linha em uma planilha Excel.
32. Crie uma função que receba um dicionário e escreva cada chave como o cabeçalho de uma coluna e os valores como as células abaixo.
33. Leia os dados de uma coluna e armazene-os em um dicionário, usando as células da primeira linha como chaves.
34. Escreva um programa que leia várias listas de um arquivo JSON e as adicione como abas separadas em um arquivo Excel.
35. Crie uma função que receba dois dicionários e escreva suas informações em duas abas diferentes.

#### **Funções e Arquivos**

36. Escreva uma função que receba o caminho de um arquivo Excel e remova todas as abas vazias.
37. Construa um script que copie o conteúdo de um arquivo Excel para outro, preservando as formatações.
38. Escreva uma função que receba o nome de uma aba e salve seu conteúdo em um arquivo separado.
39. Crie um programa que leia um arquivo Excel com dados de vendas e gere relatórios mensais em arquivos separados.
40. Desenvolva uma função que leia e combine os dados de vários arquivos Excel em um único arquivo.

#### **DataFrames e Series**

41. Leia os dados de uma planilha Excel e crie um DataFrame com eles.
42. Adicione os dados de um DataFrame a uma aba existente em um arquivo Excel.
43. Crie um programa que leia uma planilha Excel e gere estatísticas descritivas de todas as colunas numéricas.
44. Use pandas para comparar os dados de duas abas diferentes e identificar discrepâncias.
45. Escreva uma função que leia os dados de um arquivo Excel e retorne as colunas que possuem valores ausentes.

#### **Loops e Condicionais**

46. Escreva um loop que percorra todas as abas de um arquivo Excel e imprima os nomes delas.
47. Crie uma função que, para cada linha de uma planilha, imprima as células que possuem valores maiores que 100.
48. Use um loop para copiar os dados de uma coluna de um arquivo Excel para outro, ignorando células vazias.
49. Construa um script que leia dados de uma planilha, filtre apenas as linhas com valores positivos e salve-as em um novo arquivo.
50. Dado um conjunto de planilhas em um diretório, escreva um programa que verifique quais contêm uma aba chamada "Relatório".

### **Desafios Finais**

1. Desenvolva um programa que analise um arquivo Excel e destaque as células duplicadas em uma cor diferente.
2. Construa um script que leia um arquivo Excel com vendas diárias e crie uma aba separada para cada mês, com os dados correspondentes.
3. Escreva uma função que valide se os valores de uma planilha estão dentro de um intervalo especificado pelo usuário.
4. Crie um gráfico dinâmico em uma planilha que atualize automaticamente ao adicionar novos dados.
5. Desenvolva um programa que leia um arquivo Excel com dados de clientes e gere um gráfico com a distribuição de idade.
6. Escreva uma função que leia um arquivo Excel com dados financeiros e identifique as 5 maiores despesas.
7. Construa um script que leia um arquivo Excel e transforme os dados de cada aba em arquivos CSV separados.
8. Crie uma função que receba uma planilha e destaque as células com valores atípicos (outliers) em vermelho.
9. Desenvolva um programa que combine vários arquivos Excel em um só, criando abas separadas para cada arquivo.
10. Escreva um script que leia um arquivo Excel e crie um relatório em PDF com os dados resumidos.
