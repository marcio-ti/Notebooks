### **Exercícios: Formatando Dicionários em Python com `json.dumps()` e `pprint`**

#### **Básico**
1. Crie um dicionário simples com três pares chave-valor. Use `json.dumps()` para imprimir o dicionário com indentação de 2 espaços.

2. Crie um dicionário com chaves representando nomes de frutas e valores representando suas quantidades. Use `json.dumps()` para ordenar as chaves em ordem alfabética.

3. Use o módulo `json` para transformar o dicionário `{'x': 10, 'y': 20}` em uma string JSON com indentação de 4 espaços.

4. Converta o dicionário `{'a': 1, 'b': 2, 'c': 3}` para uma string JSON e imprima no console.

5. Crie um dicionário com apenas uma chave contendo uma lista como valor. Use `json.dumps()` para formatar e imprimir.

---

#### **Intermediário**
6. Dado o dicionário `{'name': 'Alice', 'age': 30, 'city': 'New York'}`, formate-o com `json.dumps()` para exibir as chaves em ordem alfabética e com indentação de 3 espaços.

7. Implemente uma função que recebe um dicionário como argumento e retorna sua versão formatada como string JSON com indentação configurável.

8. Crie um dicionário com 10 pares chave-valor. Use `json.dumps()` para formatá-lo e salve a saída em um arquivo chamado `output.json`.

9. Use o módulo `pprint` para exibir o dicionário `{'a': [1, 2, 3], 'b': [4, 5, 6]}` de forma legível.

10. Combine `json.dumps()` e o módulo `pprint` para exibir um dicionário formatado com `json.dumps()` primeiro e depois com `pprint`.

---

#### **Avançado**
11. Crie um dicionário aninhado com três níveis de profundidade e use `json.dumps()` para exibir a estrutura de forma legível.

12. Escreva um script que gera um dicionário dinâmico (com base em um loop) e o exibe ordenado com `json.dumps()`.

13. Use `json.dumps()` para exibir o dicionário `{'key': 'value', 'list': [1, 2, 3], 'nested': {'a': 10}}` com diferentes níveis de indentação (1, 2 e 4).

14. Crie um dicionário com chaves não-string (ex.: funções, números, etc.). Tente usar `json.dumps()` e explique o erro gerado.

15. Implemente uma função que aceita um dicionário e verifica se ele pode ser convertido em JSON. Se não puder, exiba uma mensagem de erro amigável.

---

#### **Desafios**
16. Use `json.dumps()` para converter o dicionário `{'students': [{'name': 'John'}, {'name': 'Jane'}]}` em uma string JSON formatada e salve em um arquivo.

17. Crie um script que lê um arquivo JSON e o converte de volta para um dicionário Python. Use `pprint` para exibir o resultado.

18. Implemente uma função que recebe uma string JSON e a converte de volta em um dicionário. Valide se a string é um JSON válido antes da conversão.

19. Escreva um programa que compara duas strings JSON e verifica se elas representam o mesmo dicionário (independentemente da ordem das chaves).

20. Crie um dicionário complexo contendo listas, dicionários aninhados e tipos diferentes. Formate-o com `json.dumps()` e `pprint`. Compare as saídas.
