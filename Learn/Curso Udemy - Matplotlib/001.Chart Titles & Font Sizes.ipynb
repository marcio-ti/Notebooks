<code><b>Estilização Avançada:</b></code>
Adicione um título ao gráfico com o texto "Gráfico Estilizado", usando uma fonte em itálico e tamanho 20.

Neste caso, para modificar o estilo da fonte, é necessária a importação de outra biblioteca

```python
from matplotlib.font_manager import FontProperties
font = FontProperties(style='italic')

x = [1,2,3,4]
y = [10,20,30,40]

fig, ax = plt.subplots()

ax.plot(x,y)
fig.suptitle('Gráfico Estilizado', fontsize=20, color='blue', fontproperties=font)

plt.show()
```

<code><b>Título com Fonte Personalizada:</b></code>
Use o método FontProperties para definir uma fonte personalizada (como "Comic Sans MS") para o título do gráfico. Escolha um tamanho de fonte de 18.

```python
from matplotlib.font_manager import FontProperties

font = FontProperties(family='Comic Sans MS', size=18)

x = [1,2,3,4]
y = [10,20,30,40]

fig, ax = plt.subplots()
ax.plot(x,y)

ax.set_title("Fonte Personalizada", fontproperties=font)

plt.show()
```

<code><b>Título com Shadow:</b></code>
Adicione um título ao gráfico com o texto "Título com Sombra", usando a propriedade shadow=True para criar uma sombra.

```python
import matplotlib.pyplot as plt
from matplotlib.patheffects import withStroke

# Dados
x = [1, 2, 3, 4]
y = [10, 20, 30, 40]

# Criar o gráfico
fig, ax = plt.subplots()
ax.plot(x, y)

# Adicionar sombra ao título
title = ax.set_title("Gráfico de Temperatura (ºC)")
title.set_path_effects([withStroke(linewidth=3, foreground="gray")])  # Sombra cinza

# Exibir o gráfico
plt.show()
```

<code><b>Título com Background:</b></code>
Crie um gráfico e adicione um título com o texto "Título com Fundo Personalizado". Defina um fundo de cor amarela (usando bbox) e bordas arredondadas.

```python
# Dados do gráfico
x = [1, 2, 3, 4]
y = [10, 20, 30, 40]

# Criar o gráfico
fig, ax = plt.subplots()
ax.plot(x, y)

# Adicionar título com fundo personalizado
ax.set_title(
	"Título com Fundo Personalizado",
	fontsize=16,
	bbox=dict(facecolor="yellow", edgecolor="black", boxstyle="round,pad=0.5")
)

# Exibir o gráfico
plt.show()
```
