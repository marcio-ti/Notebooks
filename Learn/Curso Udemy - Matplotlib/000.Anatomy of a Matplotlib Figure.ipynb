# Anatomy of a Matplotlib Figure

## PyPlot API
<hr>

```python
import matplotlib.pyplot as plt
```

```python
y = [0,1,4,9,16,25,36,49,64]
```

```python
plt.plot(y)
```

## Object-Oriented
<hr>

```python
fig = plt.figure()
ax = fig.add_subplot()

ax.plot(y)
```

Também é possível criar a Figure e o Axis separados usando a função <code>plt.subplots()</code>

```python
fig, ax = plt.subplots()

ax.plot(y)

fig.suptitle('Título Geral')
ax.set_title('Título do gráfico específico')
```

### formatting options

| Option | Object-Oriented | PyPlot API |
| --- | --- | --- |
| Figure Title | fig.suptitle() | plt.suptitle() |
| Chart Title | ax.set_title() | plt.subtitle() |
| X-Axis Label | ax.set_xlabel() | plt.xlabel() |
| Y-Axis Label | ax.set_ylabel() | plt.ylabel() |
| Legend | ax.legend() | plt.legend() |
| X-Axis Limit | ax.set_xlim() | plt.xlim() |
| Y-Axis Limit | ax.set_ylim() | plt.ylim() |
| X-Axis Ticks | ax.set_xticks() | plt.xticks() |
| Y-Axis Ticks | ax.set_yticks() | plt.yticks() |
| Vertical Line | ax.axvline() | plt.axvline() |
| Horizontal Line | ax.axhline() | plt.axhline() |
| Text | ax.text() | plt.text() |
| Spines (Borders) | ax.spines[‘side’] | plt.spines[‘side’] |

### Exemplo com a interface orientada a objetos (OO):

```python
import matplotlib.pyplot as plt

# Criando uma figura e um eixo
fig, ax = plt.subplots()

# Definindo títulos e rótulos
fig.suptitle("Título da Figura")
ax.set_title("Título do Gráfico")
ax.set_xlabel("Rótulo do Eixo X")
ax.set_ylabel("Rótulo do Eixo Y")

# Definindo limites e ticks
ax.set_xlim([0, 10])
ax.set_ylim([0, 10])
ax.set_xticks([1, 3, 5, 7, 9])
ax.set_yticks([2, 4, 6, 8])

# Adicionando uma legenda
ax.plot([1, 2, 3], [1, 2, 3], label="Linha 1")
ax.plot([1, 2, 3], [3, 2, 1], label="Linha 2")
ax.legend()

# Adicionando linhas e texto
ax.axvline(x=5, color='red')
ax.axhline(y=5, color='blue')
ax.text(2, 8, "Texto no gráfico", fontsize=12)

# Excluindo a borda superior
ax.spines['top'].set_visible(False)

# Exibindo o gráfico
plt.show()
```

### Exemplo com a API PyPlot:

```python
import matplotlib.pyplot as plt

# Definindo títulos e rótulos
plt.suptitle("Título da Figura")
plt.title("Título do Gráfico")
plt.xlabel("Rótulo do Eixo X")
plt.ylabel("Rótulo do Eixo Y")

# Criando os dados
plt.plot([1, 2, 3], [1, 2, 3], label="Linha 1")
plt.plot([1, 2, 3], [3, 2, 1], label="Linha 2")

# Definindo limites e ticks
plt.xlim([0, 10])
plt.ylim([0, 10])
plt.xticks([1, 3, 5, 7, 9])
plt.yticks([2, 4, 6, 8])

# Adicionando a legenda
plt.legend()

# Adicionando linhas e texto
plt.axvline(x=5, color='red')
plt.axhline(y=5, color='blue')
plt.text(2, 8, "Texto no gráfico", fontsize=12)

# Excluindo a borda superior
plt.gca().spines['top'].set_visible(False)

# Exibindo o gráfico
plt.show()
```
