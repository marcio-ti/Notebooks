# Font Properties

### **Exercícios para praticar o uso do `FontProperties`**

<code><b>1.</b></code> **Título Personalizado**
   Crie um gráfico simples com qualquer conjunto de dados e aplique um título estilizado usando `FontProperties` com:
   - Família: `serif`
   - Estilo: `italic`
   - Peso: `bold`
   - Cor do título: azul

```python
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties

font = FontProperties(family="serif", size=16,style="italic", weight="bold")

x = [1,3,5,3,4,5,6,7,10]
y = [2,3,3,2,3,4,5,1,9]
fig, ax = plt.subplots()
ax.plot(x,y)

ax.set_title("Testando, 1, 2, 3", fontproperties=font, color="blue")

plt.show()
```

<code><b>2.</b></code> **Rótulos Personalizados**
   Adicione rótulos personalizados para os eixos X e Y com as seguintes propriedades:
   - Família: `sans-serif`
   - Estilo: `oblique`
   - Tamanho: 14
   - Cor dos rótulos: verde



```python
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties

font = FontProperties(family="serif", size=16,style="italic", weight="bold")
estilo_rotulos = FontProperties(family="sans-serif", size=14,style="oblique")

x = [1,3,5,3,4,5,6,7,10]
y = [2,3,3,2,3,4,5,1,9]
fig, ax = plt.subplots()
ax.plot(x,y)

ax.set_title("Testando, 1, 2, 3", fontproperties=font, color="blue")
ax.set_xlabel('Eixo X', fontproperties=estilo_rotulos, color="green")
ax.set_ylabel('Eixo Y', fontproperties=estilo_rotulos, color='green')

plt.show()
```
