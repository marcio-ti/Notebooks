# Matplotlib - Tipos de Gr치ficos

## Gr치ficos de Contorno

```python
import matplotlib.pyplot as plt
import numpy as np

# Gerar dados
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

# Criar o gr치fico de contorno
plt.figure()
cp = plt.contour(X, Y, Z, levels=15, colors='black')
plt.contourf(X, Y, Z, levels=15, cmap='viridis')
plt.colorbar(cp)
plt.title('Gr치fico de Contorno')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.show()
```

```python
import matplotlib.pyplot as plt
import random
import seaborn as sns

# Definir categorias (20 itens)
categorias = [f'Cat{i}' for i in range(1, 21)]

# Gerar valores aleat칩rios (20 itens)
valores = [random.randint(-20, 20) for _ in range(20)]

# Definir as cores: azul para positivo, vermelho para negativo
cores = ['indigo' if valor >= 0 else 'crimson' for valor in valores]

fig, ax = plt.subplots(figsize=(12, 6))
barras = ax.bar(categorias, valores, color=cores)

# Adicionar r칩tulos de valores
for barra in barras:
	altura = barra.get_height()
	ax.annotate(f'{altura}',
	            xy=(barra.get_x() + barra.get_width() / 2, altura),
	            xytext=(0, 3),  # 3 points vertical offset
	            textcoords="offset points",
	            ha='center', va='bottom' if altura >= 0 else 'top')

ax.set_title('Exemplo de Gr치fico de Colunas com Valores Positivos e Negativos')
ax.set_ylabel('Valores')

# Posicionar o eixo X no ponto Y=0
ax.spines['left'].set_position(('data', 0))
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['bottom'].set_position(('data', 0))

# Adicionar linha horizontal no Y=0
ax.axhline(0, color='black', linewidth=1.3)

sns.despine(left=True, bottom=True)
plt.xticks(rotation=45)
plt.show()
```
