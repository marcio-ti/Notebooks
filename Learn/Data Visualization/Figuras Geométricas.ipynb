```python
import matplotlib.pyplot as plt

# Dados para o gráfico
x = [1, 2, 3, 4, 5]
y = [2, 3, 1, 4, 7]

# Criar a figura e os eixos
fig, ax = plt.subplots()

# Plotar os dados
ax.scatter(x, y, color='blue')

# Adicionar um retângulo
rect = plt.Rectangle((1.5, 1.5), 2, 2.5, fill=True, color='green', alpha=0.3)
ax.add_patch(rect)

# Adicionar um círculo
circle = plt.Circle((4, 5), 0.5, fill=True, color='red', alpha=0.5)
ax.add_patch(circle)

# Adicionar uma linha
line = plt.Line2D([1, 5], [2, 7], color='purple', linewidth=2, linestyle='--')
ax.add_line(line)

# Configurar o layout do gráfico
ax.set_xlabel('Eixo X')
ax.set_ylabel('Eixo Y')
ax.set_title('Gráfico com Figuras Geométricas')

# Exibir o gráfico
plt.show()
```
