```python
import matplotlib.pyplot as plt

# Dados
anos = ['2023', '2024']
valores = [100, 75]  # 2024 é menor que 2023

diff = valores[0] - valores[1]

# Criando o gráfico
fig, ax = plt.subplots()
ax.bar(anos, valores, color=['blue', 'orange'])

# Adicionando a chave para mostrar a diferença
x_chave = 1  # posição entre as barras
ax.annotate(
	'',
	xy=(x_chave, valores[0]),
	xytext=(x_chave, valores[1]),
	arrowprops=dict(arrowstyle='-[', color='red', lw=2)
)

# Adicionando o texto da diferença
ax.text(
	x_chave + 0.1, (valores[0] + valores[1]) / 2,
	f'-{diff}', color='red', fontsize=12, va='center'
)

# Rótulos
ax.set_ylabel('Valor')
ax.set_title('Comparação de 2023 e 2024')

plt.show()
```
