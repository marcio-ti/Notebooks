Aqui estão 20 exercícios para praticar os conteúdos abordados no gráfico utilizando Matplotlib e Seaborn. Eles cobrem aspectos como personalização, manipulação de dados e criação de gráficos.

### Exercícios:

1. **Gráfico de Linha Simples:**
   Crie um gráfico de linha simples utilizando `matplotlib` para representar a evolução mensal do preço de um produto ao longo do ano. Utilize um único produto como exemplo.

2. **Alteração das Cores:**
   Altere a cor da linha do gráfico para um tom diferente usando a paleta `Seaborn` ou cores customizadas.

3. **Adição de Título e Eixos:**
   Adicione um título ao gráfico, além de legendas para os eixos X e Y. Personalize o tamanho e a cor do título.

4. **Adicionando Marcações nos Eixos X e Y:**
   Crie um gráfico com dados mensais e adicione as marcações no eixo X para cada mês e no eixo Y com valores de preços.

5. **Gráfico de Linha com Várias Séries:**
   Crie um gráfico de linha para múltiplos produtos, cada um com uma linha representando seu preço ao longo do tempo. Utilize diferentes cores para cada linha.

6. **Modificação dos Ticks no Eixo X:**
   Personalize os ticks do eixo X para exibir o nome dos meses ao invés de números, usando `matplotlib`.

7. **Gráfico de Linha com Bolinhas:**
   Crie um gráfico de linha e adicione bolinhas para marcar os pontos de dados de cada mês.

8. **Texto nos Pontos de Dados:**
   Adicione texto em cada ponto do gráfico para mostrar o valor correspondente ao preço de cada mês.

9. **Gráfico com Linha Tracejada:**
   Crie um gráfico de linha e adicione linhas verticais tracejadas para separar os meses.

10. **Adicionar Anotações de Variação Percentual:**
    Calcule a variação percentual entre o primeiro e o último mês de um dos produtos e adicione essa informação no gráfico.

11. **Exclusão do Eixo Y:**
    Modifique o gráfico para ocultar completamente o eixo Y (remover marcações e linhas do eixo).

12. **Adicionando Legenda Personalizada:**
    Adicione uma legenda personalizada ao gráfico, utilizando um texto para descrever o que cada linha representa (exemplo: “Batata”, “Tomate”, “Cebola”).

13. **Gráfico de Linha com Subplots:**
    Crie um gráfico com três subgráficos (um para cada produto), organizados verticalmente. Cada subgráfico deve mostrar a evolução do preço ao longo dos meses.

14. **Gráfico com Linhas e Preenchimento:**
    Crie um gráfico de linha com o preenchimento da área abaixo da linha (área sombreada).

15. **Gráfico de Barras para Comparação de Preços:**
    Crie um gráfico de barras vertical para comparar o preço médio de cada produto durante o ano de 2024.

16. **Gráfico de Linhas com Ajustes de Posição do Texto:**
    Adicione os valores de cada ponto no gráfico, mas ajustando a posição do texto para que fique mais legível, com uma pequena margem em relação ao ponto.

17. **Gráfico de Linha com Estilo de Pontos e Linhas:**
    Alterne o estilo da linha para ser pontilhada ou pontuado e mude a forma dos marcadores de ponto.

18. **Gráfico com Diferentes Tipos de Gráficos (Mistura):**
    Crie um gráfico que combine gráficos de barras e linhas no mesmo espaço, por exemplo, utilizando uma linha para a evolução de preços e barras para representar variações mensais.

19. **Gráfico de Linha com Zoom em Intervalo de Meses:**
    Crie um gráfico com todos os meses, mas depois ajuste os limites do eixo X para mostrar apenas um intervalo específico (exemplo: março a agosto).

20. **Gráfico com Texto de Informações Adicionais:**
    Adicione texto adicional no gráfico, como por exemplo, "O preço de batata caiu drasticamente em dezembro", em um ponto específico, utilizando uma caixa de texto.

Esses exercícios cobrem um amplo espectro de funcionalidades de Matplotlib e Seaborn, desde gráficos básicos até customizações mais avançadas, permitindo uma prática abrangente desses pacotes para visualização de dados.
