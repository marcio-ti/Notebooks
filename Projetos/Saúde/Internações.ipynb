```python
import pandas as pd
```

```python
sus = pd.read_excel("sus-internaçoes.xlsx")
```

```python
df = sus.copy(deep=True)
```

```python
df[['DT_INTER','DT_SAIDA']] = df[['DT_INTER','DT_SAIDA']].astype('str')
df['DT_INTER'] = df['DT_INTER'].apply(lambda x: pd.to_datetime(f"{x[0:4]}" + "-" + f"{x[4:6]}" + f"-" + f"{x[6:8]}"))
df['DT_SAIDA'] = df['DT_SAIDA'].apply(lambda x: pd.to_datetime(f"{x[0:4]}" + "-" + f"{x[4:6]}" + f"-" + f"{x[6:8]}"))
```
