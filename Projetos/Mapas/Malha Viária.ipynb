```python
import json
import geopandas as gpd
import pandas as pd
from pandas import json_normalize
from shapely.geometry import LineString
```

```python
# Abrir e carregar o arquivo JSON
with open('canoas.json', 'r') as file:
	data = json.load(file)

# Se o JSON tiver uma lista de 'features'
features = data.get('features', [])  # Extra√≠mos a lista se ela estiver sob a chave 'features'

# Normalizar os dados para criar o DataFrame
dados = pd.json_normalize(features)

dados
```
