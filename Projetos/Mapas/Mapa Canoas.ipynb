```python
import geopandas as gpd
import pandas as pd
import folium
import matplotlib.pyplot as plt
```

```python
rs = gpd.read_file('RS/RS_Municipios_2022.shp')
```

```python
rs
```

```python
canoas = rs.copy(deep=True)
canoas = canoas.loc[canoas['NM_MUN'] == 'Canoas']
canoas.plot(figsize=(12,8), edgecolor='black', facecolor='white')
```

```python
# Selecionar uma linha específica
linha = canoas.iloc[[x for x in range(1,50)]]

fig, ax = plt.subplots(figsize=(6,6))
linha.plot(ax=ax, edgecolor='black')
plt.show()
```
