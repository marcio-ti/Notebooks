```python
def split_csv(nome_arquivo, tamanho_arquivos):
	try:
		with open(nome_arquivo, 'r', newline='', encoding='utf-8') as csvfile:
			reader = csv.reader(csvfile)
			headers = next(reader)
			headers.append('nome_arquivo')  # Adiciona a coluna do nome do arquivo de sa√≠da
			bloco = 1
			linhas = []

			for i, row in enumerate(reader, start=1):
				linhas.append(row)
				if i % tamanho_arquivos == 0:
					write_chunk(headers, linhas, bloco)
					bloco += 1
					linhas = []

			if linhas:
				write_chunk(headers, linhas, bloco)

		print("Arquivo dividido com sucesso!")

	except Exception as e:
		print(f"Erro ao processar o arquivo: {e}")

def write_chunk(headers, linhas, bloco):
	chunk_nome_arquivo = f"{pasta_destino}\\output_{bloco:02d}.csv"
	try:
		with open(chunk_nome_arquivo, 'w', newline='', encoding='utf-8') as csvfile:
			writer = csv.writer(csvfile)
			writer.writerow(headers)
			for linha in linhas:
				linha.append(chunk_nome_arquivo)  # Adiciona o nome do arquivo de sa√≠da em cada linha
				writer.writerow(linha)
		print(f"{chunk_nome_arquivo} criado com sucesso!")
	except Exception as e:
		print(f"Erro ao escrever o arquivo {chunk_nome_arquivo}: {e}")

# Use o nome do seu arquivo CSV aqui
nome_arquivo = arquivo_csv

tamanho_arquivos = 15000

split_csv(nome_arquivo, tamanho_arquivos)
```

### Ajustando os dados dos arquivos conforme o padr√£o UTF-8

```python

ajustes_textos = {
	'√É∆ë': 'A',
	'√É∆í':'A',
	'¬Å':'A',
	'√É':'A',
	'‚Ä°':'C',
	'√á':'C',
	'¬Ä':'C',
	'‚Ç¨':'C'

}

def substituir_caracteres(texto, ajustes):
	for chave, valor in ajustes.items():
		texto = re.sub(chave, valor, texto)
	return texto

def verificar_arquivos_csv(diretorio, ajustes):
	for arquivo in os.listdir(diretorio):
		if arquivo.endswith(".csv"):
			caminho_arquivo = os.path.join(diretorio, arquivo)
			df = pd.read_csv(caminho_arquivo)
			df = df.map(lambda x: substituir_caracteres(str(x), ajustes) if isinstance(x, str) else x)
			df.to_csv(caminho_arquivo, index=False)  # Salva as mudan√ßas no pr√≥prio arquivo
		print(f"{arquivo} Conclu√≠do!")

# Diret√≥rio com os arquivos CSV
diretorio_arquivos = pasta_destino

# Verificar os arquivos
verificar_arquivos_csv(diretorio_arquivos, ajustes_textos)
```
