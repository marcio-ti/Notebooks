```python
import json

import pandas as pd
```

```python
lista_ceasa_online = {}
caminho = r'C:\Users\divtec-dados\DataspellProjects\Notebooks\Ceasa\Dados\JSON\lista_nomes_produtos_ceasa_online.json'

with open(caminho, encoding='utf-8') as arquivo_json:
	dados = json.load(arquivo_json)
	lista_ceasa_online.update(dados)


def ajuste_produto(x):
	return lista_ceasa_online.get(x, x)


ajuste_unidade = {
	0: "KG",
	1: "CX",
	2: "SACO",
	3: "ENG",
	4: "UND",
	5: "DZ",
	6: "CXT",
	7: "SCL",
	8: "MOL",
	9: "OUTROS"
}
```

### Notas Fiscais 2018 a 2022
---

```python
df_2018_2022 = pd.read_csv(
	r'C:\Users\divtec-dados\DataspellProjects\Notebooks\Ceasa\Dados\divtec_old_notas_fiscais_copia.csv')
df_2018_2022['data'] = df_2018_2022['data'].astype('datetime64[ns]')
df_2018_2022['produto'] = df_2018_2022['produto'].apply(lambda texto: str.strip(texto))
df_2018_2022['ano'] = df_2018_2022['data'].dt.year
df_2018_2022['mes'] = df_2018_2022['data'].dt.month
df_2018_2022 = df_2018_2022[
	['ano', 'mes', 'data',
	 'nota_fiscal', 'procedencia', 'destino',
	 'matricula', 'produto', 'unidade_str', 'quantidade'
	 ]]
df_2018_2022
```

```python
df_2018_2022['produto'] = df_2018_2022['produto'].apply(ajuste_produto)
df_2018_2022.rename(columns={'unidade_str': 'unidade'}, inplace=True)
df_2018_2022['unidade'] = df_2018_2022['unidade'].apply(lambda texto: str.strip(texto))
df_2018_2022 = pd.merge(df_2018_2022, pesos_ceasa_online, how='left', on=['produto', 'unidade'])
df_2018_2022
```

```python
erros = df_2018_2022[df_2018_2022['codigo'].isna()]
erros = erros[['produto', 'unidade']].drop_duplicates()
erros = erros.sort_values(by=['produto', 'unidade']).sort_values('produto')
erros
```

```python
erros_rows = df_2018_2022[df_2018_2022['codigo'].isna()]
erros_rows
```
