```python
%%sql
SELECT
	destino,
	procedencia,
	cad_cidades.coredes_rs,
	produto,
	SUM(volume) as volume
FROM
	divtec_notasfiscais
	LEFT JOIN
		cad_cidades ON divtec_notasfiscais.procedencia = cad_cidades.cidade
	LEFT JOIN
		cad_usuarios ON divtec_notasfiscais.destino = cad_usuarios.matricula_nome
WHERE
	tipo = 'PRODUTOR'
GROUP BY
	destino,
	procedencia,
	cad_cidades.coredes_rs,
	produto
ORDER BY
	cad_cidades.coredes_rs,
	destino,
	procedencia,
	produto;
```
