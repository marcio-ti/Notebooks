```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
pd.options.display.float_format = '{:.2f}'.format
import datetime as dt
```

```python
dados = pd.read_csv(r"../dados/dados2023 sem matricula.csv", sep=';', decimal=',', thousands='.')
dados['data'] = dados['data'].astype('datetime64[ns]')
dados
```

```python
produtos_grupos = pd.read_csv('../dados/produtos.csv')
```

```python
produtos_grupos = produtos_grupos[['produto','grupo','categoria','setor']]
```

```python
df = pd.merge(left=dados, right=produtos_grupos, how='left', on='produto')
```

```python
df
```

```python
df['mes'] = df['data'].dt.month
df = df[['mes','produto','categoria','grupo','setor','volume','valor comercializado','cotação','procedencia','estado','atac_gnp']]
```

```python
df_cotacao = df.groupby(['mes','categoria','produto'])[['cotação']].mean().reset_index()
df_cotacao = df_cotacao.groupby(['categoria','produto'])[['cotação']].mean().reset_index()

df_valor_volume = df.groupby(['mes','categoria','produto','procedencia','estado','atac_gnp'])[['volume','valor comercializado']].sum().reset_index()
```

```python
df_final = pd.merge(left=df_valor_volume, right=df_cotacao, how='inner', on=['categoria','produto'])
```
