```python
def gerar_grafico_selecionados(produto01, produto02,produto03, produto04, df):
    cor_da_media='crimson'
    # Alterar o nome dos produtos
    df['PRODUTO'] = df['PRODUTO'].replace('08901-OVO BRANCO CX30DZ','08901-OVO BRANCO', regex=False)
    df['PRODUTO'] = df['PRODUTO'].replace('08902-OVO VERMELHO CX30DZ','08902-OVO VERMELHO', regex=False)
    
    # ------------------------------------------------------------------------------------------------------------------------------ #
    # Filtrando o dataframe pelos produtos selecionados
    produto01 = df.loc[df['PRODUTO'] == produto01]
    produto01['PRODUTO'] = produto01['PRODUTO'].apply(lambda x: str(x[6:]).title())
    
    produto02 = df.loc[df['PRODUTO'] == produto02]
    produto02['PRODUTO'] = produto02['PRODUTO'].apply(lambda x: str(x[6:]).title())
    
    produto03 = df.loc[df['PRODUTO'] == produto03]
    produto03['PRODUTO'] = produto03['PRODUTO'].apply(lambda x: str(x[6:]).title())
    
    produto04 = df.loc[df['PRODUTO'] == produto04]
    produto04['PRODUTO'] = produto04['PRODUTO'].apply(lambda x: str(x[6:]).title())
    
    # ------------------------------------------------------------------------------------------------------------------------------ #
    # Crie os subplots
    fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(20, 8))
    sns.set_theme(style='white')
    
    # ------------------------------------------------------------------------------------------------------------------------------ #   
    # Gráfico 1
    axs[0, 0] = sns.lineplot(data=produto01, x='DATA', y='PRECO', ax=axs[0,0], color=CINZA2, linewidth=2)
    axs[0,0].scatter(produto01['DATA'], produto01['PRECO'], color=COMPLEMENTAR1, label='_nolegend_')
    for i, row in produto01.iterrows():
        preco = str(round(row['PRECO'],2)).replace(',','#').replace('.',',').replace('#','.')
        axs[0,0].annotate(f"{preco}", (row['DATA'], row['PRECO']), textcoords="offset points", xytext=(0,10), ha='center')
    axs[0,0].set_ylabel("Preço")
    axs[0,0].set_xlabel(f"{produto01['PRODUTO'].iloc[0]} ({produto01['UNIDADE'].iloc[0]})", fontsize=16)
    axs[0,0].grid(False)
    axs[0,0].axhline(y=produto01['PRECO'].mean(), color=cor_da_media, linestyle='--', label=f"Média - {round(produto01['PRECO'].mean(),2)}")
    axs[0,0].xaxis.set_major_formatter(mdates.DateFormatter('%d/%m')) # Formata as datas no eixo x
    axs[0,0].set_xticks(produto01['DATA'].unique())  # Define as datas que devem ser exibidas no eixo x
    axs[0,0].legend(loc='best')
    
    # ------------------------------------------------------------------------------------------------------------------------------ #
    # Gráfico 2
    axs[0, 1] = sns.lineplot(data=produto02, x='DATA', y='PRECO', ax=axs[0,1], color=CINZA2, linewidth=2)
    axs[0,1].scatter(produto02['DATA'], produto02['PRECO'], color=COMPLEMENTAR1, label='_nolegend_')
    for i, row in produto02.iterrows():
        preco = str(round(row['PRECO'],2)).replace(',','#').replace('.',',').replace('#','.')
        axs[0,1].annotate(f"{preco}", (row['DATA'], row['PRECO']), textcoords="offset points", xytext=(0,10), ha='center')
    axs[0,1].set_ylabel("Preço")
    axs[0,1].set_xlabel(f"{produto02['PRODUTO'].iloc[0]} ({produto02['UNIDADE'].iloc[0]})", fontsize=16)
    axs[0,1].grid(False)
    axs[0,1].axhline(y=produto02['PRECO'].mean(), color=cor_da_media, linestyle='--', label=f"Média - {round(produto02['PRECO'].mean(),2)}")
    axs[0,1].xaxis.set_major_formatter(mdates.DateFormatter('%d/%m')) # Formata as datas no eixo x
    axs[0,1].set_xticks(produto02['DATA'].unique())  # Define as datas que devem ser exibidas no eixo x
    axs[0,1].legend(loc='best')
    
    # ------------------------------------------------------------------------------------------------------------------------------ #
    # Gráfico 3
    axs[1, 0] = sns.lineplot(data=produto03, x='DATA', y='PRECO', ax=axs[1, 0], color=CINZA2, linewidth=2)
    axs[1,0].scatter(produto03['DATA'], produto03['PRECO'], color=COMPLEMENTAR1, label='_nolegend_')
    for i, row in produto03.iterrows():
        preco = str(round(row['PRECO'],2)).replace(',','#').replace('.',',').replace('#','.')
        axs[1,0].annotate(f"{preco}", (row['DATA'], row['PRECO']), textcoords="offset points", xytext=(0,10), ha='center')
    axs[1, 0].set_ylabel("Preço")
    axs[1, 0].set_xlabel(f"{produto03['PRODUTO'].iloc[0]} ({produto03['UNIDADE'].iloc[0]})", fontsize=16)
    axs[1, 0].grid(False)
    axs[1, 0].axhline(y=produto03['PRECO'].mean(), color=cor_da_media, linestyle='--', label=f"Média - {round(produto03['PRECO'].mean(),2)}")
    axs[1, 0].xaxis.set_major_formatter(mdates.DateFormatter('%d/%m')) # Formata as datas no eixo x
    axs[1, 0].set_xticks(produto03['DATA'].unique())  # Define as datas que devem ser exibidas no eixo x
    axs[1, 0].legend(loc='best')

    # ------------------------------------------------------------------------------------------------------------------------------ #
    # Gráfico 4
    axs[1, 1] = sns.lineplot(data=produto04, x='DATA', y='PRECO', ax=axs[1, 1], color=CINZA2, linewidth=2)
    axs[1,1].scatter(produto04['DATA'], produto04['PRECO'], color=COMPLEMENTAR1, label='_nolegend_')
    for i, row in produto04.iterrows():
        preco = str(round(row['PRECO'],2)).replace(',','#').replace('.',',').replace('#','.')
        axs[1,1].annotate(f"{preco}", (row['DATA'], row['PRECO']), textcoords="offset points", xytext=(0,10), ha='center')
    axs[1, 1].set_ylabel("Preço")
    axs[1, 1].set_xlabel(f"{produto04['PRODUTO'].iloc[0]} ({produto04['UNIDADE'].iloc[0]})", fontsize=16)
    axs[1, 1].grid(False)
    mean_value04 = produto04['PRECO'].mean()
    axs[1, 1].axhline(y=mean_value04, color=cor_da_media, linestyle='--', label=f"Média - {round(produto04['PRECO'].mean(),2)}")
    axs[1, 1].xaxis.set_major_formatter(mdates.DateFormatter('%d/%m'))  # Formata as datas no eixo x
    axs[1, 1].set_xticks(produto04['DATA'].unique())  # Define as datas que devem ser exibidas no eixo x
    axs[1, 1].legend(loc='best')
    
    # Remova as bordas desnecessárias
    sns.despine()
    
    # Ajuste o layout
    fig.tight_layout()
    
    # Retorno do gráfico
    return fig

def gerar_grafico_selecionados_color(produto01, produto02,produto03, produto04, df):
    # Alterar o nome dos produtos
    df['PRODUTO'] = df['PRODUTO'].replace('08901-OVO BRANCO CX30DZ','08901-OVO BRANCO', regex=False)
    df['PRODUTO'] = df['PRODUTO'].replace('08902-OVO VERMELHO CX30DZ','08902-OVO VERMELHO', regex=False)
    
    # ------------------------------------------------------------------------------------------------------------------------------ #
    # Filtrando o dataframe pelos produtos selecionados
    produto01 = df.loc[df['PRODUTO'] == produto01]
    produto01['PRODUTO'] = produto01['PRODUTO'].apply(lambda x: str(x[6:]).title())
    
    produto02 = df.loc[df['PRODUTO'] == produto02]
    produto02['PRODUTO'] = produto02['PRODUTO'].apply(lambda x: str(x[6:]).title())
    
    produto03 = df.loc[df['PRODUTO'] == produto03]
    produto03['PRODUTO'] = produto03['PRODUTO'].apply(lambda x: str(x[6:]).title())
    
    produto04 = df.loc[df['PRODUTO'] == produto04]
    produto04['PRODUTO'] = produto04['PRODUTO'].apply(lambda x: str(x[6:]).title())
    
    # ------------------------------------------------------------------------------------------------------------------------------ #
    # Crie os subplots
    fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(20, 8))
    
    
    # ------------------------------------------------------------------------------------------------------------------------------ #   
    # Gráfico 1
    axs[0, 0] = sns.lineplot(data=produto01, x='DATA', y='PRECO', ax=axs[0, 0], color=CINZA2, linewidth=2)
    axs[0, 0].scatter(produto01['DATA'], produto01['PRECO'], color=COMPLEMENTAR1, label='_nolegend_')
    
    # Anotações
    for i, row in produto01.iterrows():
        axs[0, 0].annotate(f"{row['PRECO']:.2f}", (row['DATA'], row['PRECO']), textcoords="offset points", xytext=(0, 10), ha='center')
    
    # Configurações do gráfico
    axs[0, 0].set_ylabel("Preço")
    axs[0, 0].set_xlabel(f"{produto01['PRODUTO'].iloc[0]} ({produto01['UNIDADE'].iloc[0]})", fontsize=16)
    axs[0, 0].grid(False)
    axs[0, 0].xaxis.set_major_formatter(mdates.DateFormatter('%d/%m'))
    axs[0, 0].set_xticks(produto01['DATA'].unique())
    
    # Linha da média
    media_preco = produto01['PRECO'].mean()
    axs[0, 0].axhline(y=media_preco, color=cor_da_media, linestyle='--', label=f"Média - R$ {round(media_preco, 2)}")
    
    # Alterando a cor da linha com base no valor do preço
    for i in range(len(produto01) - 1):
        if produto01['PRECO'].iloc[i] > media_preco:
            axs[0, 0].plot([produto01['DATA'].iloc[i], produto01['DATA'].iloc[i + 1]],
                           [produto01['PRECO'].iloc[i], produto01['PRECO'].iloc[i + 1]], color='green', linewidth=2)
        else:
            axs[0, 0].plot([produto01['DATA'].iloc[i], produto01['DATA'].iloc[i + 1]],
                           [produto01['PRECO'].iloc[i], produto01['PRECO'].iloc[i + 1]], color='red', linewidth=2)
    
    # Legenda
    axs[0, 0].legend(loc='best')
    
    # ------------------------------------------------------------------------------------------------------------------------------ #
    # Gráfico 2
    axs[0, 1] = sns.lineplot(data=produto02, x='DATA', y='PRECO', ax=axs[0, 1], color=CINZA2, linewidth=2)
    axs[0, 1].scatter(produto01['DATA'], produto02['PRECO'], color=COMPLEMENTAR1, label='_nolegend_')
    
    # Anotações
    for i, row in produto02.iterrows():
        axs[0, 1].annotate(f"{row['PRECO']:.2f}", (row['DATA'], row['PRECO']), textcoords="offset points", xytext=(0, 10), ha='center')
    
    # Configurações do gráfico
    axs[0, 1].set_ylabel("Preço")
    axs[0, 1].set_xlabel(f"{produto02['PRODUTO'].iloc[0]} ({produto02['UNIDADE'].iloc[0]})", fontsize=16)
    axs[0, 1].grid(False)
    axs[0, 1].xaxis.set_major_formatter(mdates.DateFormatter('%d/%m'))
    axs[0, 1].set_xticks(produto02['DATA'].unique())
    
    # Linha da média
    media_preco = produto02['PRECO'].mean()
    axs[0, 1].axhline(y=media_preco, color=cor_da_media, linestyle='--', label=f"Média - R$ {round(media_preco, 2)}")
    
    # Alterando a cor da linha com base no valor do preço
    for i in range(len(produto02) - 1):
        if produto02['PRECO'].iloc[i] > media_preco:
            axs[0, 1].plot([produto02['DATA'].iloc[i], produto02['DATA'].iloc[i + 1]],
                           [produto02['PRECO'].iloc[i], produto02['PRECO'].iloc[i + 1]], color='green', linewidth=2)
        else:
            axs[0, 1].plot([produto02['DATA'].iloc[i], produto02['DATA'].iloc[i + 1]],
                           [produto02['PRECO'].iloc[i], produto02['PRECO'].iloc[i + 1]], color='red', linewidth=2)
    
    # Legenda
    axs[0, 1].legend(loc='best')
    
    # ------------------------------------------------------------------------------------------------------------------------------ #
    # Gráfico 3
    axs[1, 0] = sns.lineplot(data=produto03, x='DATA', y='PRECO', ax=axs[1, 0], color=CINZA2, linewidth=2)
    axs[1,0].scatter(produto03['DATA'], produto03['PRECO'], color=COMPLEMENTAR1, label='_nolegend_')
    for i, row in produto03.iterrows():
        axs[1, 0].annotate(f"{row['PRECO']:.2f}", (row['DATA'], row['PRECO']), textcoords="offset points", xytext=(0,10), ha='center')
    axs[1, 0].set_ylabel("Preço")
    axs[1, 0].set_xlabel(f"{produto03['PRODUTO'].iloc[0]} ({produto03['UNIDADE'].iloc[0]})", fontsize=16)
    axs[1, 0].grid(False)
    axs[1, 0].axhline(y=produto03['PRECO'].mean(), color=cor_da_media, linestyle='--', label=f"Média - R$ {round(produto03['PRECO'].mean(),2)}")
    axs[1, 0].xaxis.set_major_formatter(mdates.DateFormatter('%d/%m')) # Formata as datas no eixo x
    axs[1, 0].set_xticks(produto03['DATA'].unique())  # Define as datas que devem ser exibidas no eixo x
    axs[1, 0].legend(loc='best')

    # ------------------------------------------------------------------------------------------------------------------------------ #
    # Gráfico 4
    axs[1, 1] = sns.lineplot(data=produto04, x='DATA', y='PRECO', ax=axs[1, 1], color=CINZA2, linewidth=2)
    axs[1,1].scatter(produto04['DATA'], produto04['PRECO'], color=COMPLEMENTAR1, label='_nolegend_')
    for i, row in produto04.iterrows():
        axs[1, 1].annotate(f"{row['PRECO']:.2f}", (row['DATA'], row['PRECO']), textcoords="offset points", xytext=(0,10), ha='center')
    axs[1, 1].set_ylabel("Preço")
    axs[1, 1].set_xlabel(f"{produto04['PRODUTO'].iloc[0]} ({produto04['UNIDADE'].iloc[0]})", fontsize=16)
    axs[1, 1].grid(False)
    mean_value04 = produto04['PRECO'].mean()
    axs[1, 1].axhline(y=mean_value04, color=cor_da_media, linestyle='--', label=f"Média - R$ {round(produto04['PRECO'].mean(),2)}")
    axs[1, 1].xaxis.set_major_formatter(mdates.DateFormatter('%d/%m'))  # Formata as datas no eixo x
    axs[1, 1].set_xticks(produto04['DATA'].unique())  # Define as datas que devem ser exibidas no eixo x
    axs[1, 1].legend(loc='best')
    
    # Remova as bordas desnecessárias
    sns.despine()
    
    # Ajuste o layout
    fig.tight_layout()
    
    # Retorno do gráfico
    return fig

def cotacao(produto, midia, titulo_grafico, ax_min=0, ay_min=0, ax_max=0, ay_max=0):
    if midia == 'instagram':
        width_midia = 1000
        height_midia = 300
        source_y = 1.4
        source_x = 0.85
        title_y = 1.4
    elif midia == 'powerpoint':
        width_midia = 1200
        height_midia = 400
        source_y = 1.17
        source_x = 0.80
        title_y = 1.20

    df_produto = dados[dados['PRODUTO'] == produto].copy(deep=True)
    # Converter a coluna 'DATA' para o tipo datetime
    df_produto['DATA'] = pd.to_datetime(df_produto['DATA'])
    # Formatar a data para o formato 'dia-mês'
    df_produto['DATA'] = df_produto['DATA'].dt.strftime('%d-%B')
    # Substituir os nomes dos meses em inglês pelos nomes dos meses em português
    df_produto['DATA'] = df_produto['DATA'].replace({'-January': '-Jan', '-February': '-Fev', '-March': '-Mar',
                                                      '-April': '-Abr', '-May': '-Mai', '-June': '-Jun',
                                                      '-July': '-Jul', '-August': '-Ago', '-September': '-Set',
                                                      '-October': '-Out', '-November': '-Nov', '-December': '-Dez'}, regex=True)

    

    if df_produto['PRECO'].iloc[0] > df_produto['PRECO'].iloc[-1]:
        cor_valor = COMPLEMENTAR1
    elif df_produto['PRECO'].iloc[0] < df_produto['PRECO'].iloc[-1]:
        cor_valor = COMPLEMENTAR1
    else:
        cor_valor = COMPLEMENTAR1

    fig = go.Figure()

    fig.add_trace(
        go.Scatter(
            x=df_produto['DATA'],
            y=df_produto['PRECO'],
            mode='lines+markers',
            line=dict(color='rgb(0, 0, 0)', width=2),  # Use uma cor de linha mais escura
            marker=dict(size=5, color='rgb(0, 0, 0)', line=dict(color='rgb(0, 0, 0)', width=2))
        )
    )
    fig.update_layout(
        width=width_midia,
        height=height_midia,
        xaxis=dict(
            showline=True,
            showgrid=False,
            showticklabels=True,
            linecolor='rgb(204, 204, 204)',
            linewidth=2,
            ticks='outside',
            tickfont=dict(
                family='Arial',
                size=12,
                color='rgb(82, 82, 82)',
            ),
            tickangle=-45  # Rotate x-axis labels for better readability
        ),
        yaxis=dict(
            showgrid=False,
            zeroline=False,
            showline=False,
            showticklabels=False,
        ),
        autosize=False,
        margin=dict(
            autoexpand=False,
            l=100,
            r=20,
            t=110,
            b=50  # Add some space at the bottom for the source annotation
        ),
        showlegend=False,
        plot_bgcolor='white',
        paper_bgcolor='white'  # Set the background color of the plot area
    )

    # Add annotations
    annotations = []

    # Labeling the left side of the plot
    annotations.append(dict(xref='paper', x=0.05, y=df_produto['PRECO'].iloc[0],
                            xanchor='right', yanchor='middle',
                            text=f"R$ {df_produto['PRECO'].iloc[0]:.2f}",
                            font=dict(family='Arial', size=20, weight='bold'),
                            showarrow=False))

    # Labeling the right side of the plot
    annotations.append(dict(xref='paper', x=0.95, y=df_produto['PRECO'].iloc[-1],
                            xanchor='left', yanchor='middle',
                            text=f"R$ {df_produto['PRECO'].iloc[-1]:.2f}",
                            font=dict(family='Arial', size=20, color=cor_valor, weight='bold'),
                            showarrow=False))

    # Title
    annotations.append(dict(xref='paper', yref='paper', x=0.0, y=title_y,
                            xanchor='left', yanchor='bottom',
                            text=f'Cotação - {titulo_grafico} ({df_produto["UNIDADE"].iloc[0].lower()})',
                            font=dict(family='Arial', size=22, color='rgb(37,37,37)'),
                            showarrow=False))

    # Source
    annotations.append(dict(xref='paper', yref='paper', x=source_x, y=source_y,
                            xanchor='center', yanchor='bottom',
                            text='CEASA RS - Gerência Técnica',
                            font=dict(family='Arial', size=11, color='rgb(150,150,150)'),
                            showarrow=False))

    # Encontrar o índice do valor máximo e mínimo
    max_index = df_produto['PRECO'].idxmax()
    min_index = df_produto['PRECO'].idxmin()

    # Adicionar anotação no ponto máximo
    annotations.append(dict(xref='x', yref='y', x=df_produto['DATA'][max_index],
                            y=df_produto['PRECO'][max_index],
                            xanchor='center', yanchor='bottom',
                            text=f"Máximo: R$ {df_produto['PRECO'][max_index]:.2f}",
                            font=dict(family='Arial', size=14, color='rgb(255, 0, 0)'),  # Use uma cor mais brilhante
                            showarrow=True, arrowhead=7, arrowsize=1.5, arrowwidth=2, arrowcolor='rgb(255, 0, 0)',ax = ax_max, ay=ay_max)) 

    # Verificar se a anotação do valor máximo está sobrepondo a linha do gráfico
    if abs(df_produto['PRECO'][max_index] - df_produto['PRECO'][min_index]) < 0.1 * df_produto['PRECO'][min_index]:
        yanchor_max = 'top'
    else:
        yanchor_max = 'bottom'

    # Atualizar anotação no ponto máximo
    annotations[-1]['yanchor'] = yanchor_max

    # Adicionar anotação no ponto mínimo
    annotations.append(dict(xref='x', yref='y', x=df_produto['DATA'][min_index],
                            y=df_produto['PRECO'][min_index],
                            xanchor='auto', yanchor='bottom',
                            text=f"Mínimo: R$ {df_produto['PRECO'][min_index]:.2f}",
                            font=dict(family='Arial', size=14, color=COMPLEMENTAR1),  # Use uma cor mais brilhante
                            showarrow=True, arrowhead=7, arrowsize=1.5, arrowwidth=2, arrowcolor=COMPLEMENTAR1,ax = ax_min, ay=ay_min)) 

    # Verificar se a anotação do valor mínimo está sobrepondo a linha do gráfico
    if abs(df_produto['PRECO'][min_index] - df_produto['PRECO'][max_index]) < 0.1 * df_produto['PRECO'][max_index]:
        yanchor_min = 'top'
    else:
        yanchor_min = 'bottom'

    # Atualizar anotação no ponto mínimo
    annotations[-1]['yanchor'] = yanchor_min

    fig.update_layout(annotations=annotations)

    return fig

def cotacao_bar(produto, titulo_grafico):
    df_produto = dados_cotacoes[dados_cotacoes['PRODUTO'] == produto].copy(deep=True)

    fig = go.Figure()

    # Encontrar o índice do valor máximo e mínimo
    max_index = df_produto['PRECO'].idxmax()
    min_index = df_produto['PRECO'].idxmin()
    
    # Criar uma lista de cores para as barras
    colors = [CINZA4 if (x != max_index and x != min_index) else 'red' if x == max_index else VERDE3 for x in df_produto.index]
    
    fig.add_trace(
        go.Bar(
            x=df_produto['DATA'],
            y=df_produto['PRECO'],
            marker=dict(color=colors),
        )
    )

    fig.add_trace(
        go.Scatter(
            x=df_produto['DATA'],
            y=[df_produto['PRECO'].mean() for x in range(len(df_produto['PRECO']))],
            mode='lines+markers',
            line=dict(color=COMPLEMENTAR5, width=3),
        )
    )
    fig.update_layout(
        xaxis=dict(
            showline=True,
            showgrid=False,
            showticklabels=True,
            linecolor='rgb(204, 204, 204)',
            linewidth=2,
            ticks='outside',
            tickfont=dict(
                family='Arial',
                size=12,
                color='rgb(82, 82, 82)',
            ),
            tickangle=-45  # Rotate x-axis labels for better readability
        ),
        yaxis=dict(
            showgrid=False,
            zeroline=False,
            showline=False,
            showticklabels=False,
        ),
        autosize=False,
        margin=dict(
            autoexpand=False,
            l=100,
            r=20,
            t=110,
            b=50  # Add some space at the bottom for the source annotation
        ),
        showlegend=False,
        plot_bgcolor='white',
        paper_bgcolor='white'  # Set the background color of the plot area
    )
    
    return fig
```
