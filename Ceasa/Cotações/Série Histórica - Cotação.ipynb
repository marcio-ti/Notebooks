```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns
import datetime
import warnings
warnings.filterwarnings('ignore')
```

## Gráfico

```python
datas = pd.date_range(start='2021-01-01', periods=104, freq='W')

# Função auxiliar para criar dados de um produto
def criar_dados_produto(codigo_nome, unidade):
    return pd.DataFrame({
        'DATA': datas,
        'PRODUTO': [codigo_nome] * len(datas),
        'UNIDADE': [unidade] * len(datas),
        'PRECO': np.round(np.random.uniform(5.0, 7.0, size=len(datas)), 2),
        'MAXIMO': np.round(np.random.uniform(7.0, 10.0, size=len(datas)), 2),
        'MINIMO': np.round(np.random.uniform(2.0, 5.0, size=len(datas)), 2),
    })

# Criação dos quatro produtos fictícios
df = criar_dados_produto('CENOURA', 'KG')

df
```

Teste da criação do Gráfico

```python
grafico = sns.lineplot(df )

grafico.plot();
```
